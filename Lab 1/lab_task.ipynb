### Algorithm: Bisection Method

1. **Start**

2. **Define** the function `f(x)`

3. **Choose** initial guesses `a` and `b` such that  
   `f(a) * f(b) < 0`

4. **Choose** a pre-specified tolerable error `e`

5. **Calculate** the new approximated root:  
   `c = (a + b) / 2`

6. **Evaluate** `f(a) * f(c)`:

   - a. If `f(a) * f(c) < 0`, then set `a = a`, `b = c`

   - b. If `f(a) * f(c) > 0`, then set `a = c`, `b = b`

   - c. If `f(a) * f(c) = 0`, then go to step 8

7. If `|f(c)| > e`, go to step 5;  
   otherwise, go to step 8

8. **Display** `c` as the root

9. **Stop**


```python
### Python Program: Bisection Method
import matplotlib.pyplot as plt

def f(x):
    # Define the function here
    return x**3 - 4*x - 9

def bisection(a, b, tol):
    if f(a) * f(b) >= 0:
        print("Bisection method fails. Choose a different interval [a, b].")
        return None

    print(f"{'Iter':<5} {'a':<10} {'b':<10} {'c':<10} {'f(c)':<15}")
    print("-" * 55)
    iterations = []
    approximations = []

    iteration = 1
    while True:
        c = (a + b) / 2.0
        iterations.append(iteration)
        approximations.append(c)

        fc = f(c)
        print(f"{iteration:<5} {a:<10.6f} {b:<10.6f} {c:<10.6f} {fc:<15.10f}")
        
        if abs(fc) < tol:
            break  # Root found
        elif f(a) * fc < 0:
            b = c
        else:
            a = c

        iteration += 1

    print("-" * 50)
    plt.plot(iterations, approximations, marker='')
    plt.title("Convergence of Root Approximation (Bisection Method)")
    plt.xlabel("Iteration")
    plt.ylabel("Approximate Root (c)")
    plt.grid(True)
    plt.show()

    return c

# User Input
a = float(input("Enter the first initial value a: "))
b = float(input("Enter the second initial value b: "))
tolerance = float(input("Enter the tolerance (e.g. 0.00001): "))

root = bisection(a, b, tolerance)
if root is not None:
    print(f"\nThe root is approximately: {root:.6f}")
```
